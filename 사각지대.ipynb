- 노인 교통사고 건수(자치구별),  노인 보호 구역 

```python
import requests
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import os
import webbrowser
import folium
from folium import plugins
print(folium.__version__)
```

```python
import json
geo_path = 'data\\02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
prot_df = pd.read_csv('data\\seoul_protection_facility.csv', index_col = 0)
accident_df =  pd.read_csv('data\\total.csv',header=None)
```

```python
prot_df
```

```python
accident_df
```

```python
try:
    accident_df = accident_df.drop(0 ,axis=0)
    accident_df = accident_df.drop(1 ,axis=0)
except:
    pass
```

```python
accident_df = accident_df.set_index(7)
```

```python
accident_df
```

```python
df_total = pd.concat([accident_df, prot_df],  axis = 1)[[6,'노인','어린이']]
df_total
```

```python
#보호시설이 없는 경우는 최대치 넣음
```

```python
df_ratio = df_total[6].astype(int).div(df_total['노인']).replace([np.inf, -np.inf],0)
```

```python
df_ratio = df_ratio.to_frame('ratio')
```

```python
df_ratio
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11,
                 tiles='Stamen Toner')
map.choropleth(geo_data = geo_str,
               data = df_ratio['ratio'],
               columns = [df_ratio.index, df_ratio['ratio']],
               fill_color = 'PuRd', #puRd, YlGnBu
               key_on = 'feature.id')
```
