# 열 추가하기

```python
ratio = []

for i in range(len(old)):
    a = float(old[i].replace(',',''))
    b = float(total[i].replace(',',''))
    ratio.append(a/b)
    
print(ratio)
```

```python
peoplesum['노인비율'] = ratio
ratio
```
